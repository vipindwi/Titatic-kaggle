```python
import matplotlib.pyplot as plt
%matplotlib inline
import numpy as np
import pandas
from sklearn.svm import SVC
from sklearn.feature_selection import RFE
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, precision_recall_curve
import csv
import math
```

```python
df  = pandas.read_csv('train.csv')
```

```python
df = df.drop(['Cabin','Embarked'], axis=1)
```

```python
df.dropna()
```

```python
labels = []
features = []
for index, row in df.iterrows():
    X = []
    X.append(row['PassengerId'])
    X.append(row['Pclass'])
    
    if row['Sex'] == 'male':
        X.append(0)
    else:
        X.append(1)
    X.append(row['SibSp'])
    X.append(row['Parch'])
    
    va = int (row['Fare'])
    X.append(row['Fare'])
    
    features.append(X)
    labels.append(row['Survived'])
    
```

```python
num = len(features)
split = int(0.7 * num)

labels_train = np.array(labels[:split]).astype(np.float64)
features_train = np.array(features[:split]).astype(np.float64)

labels_test = np.array(labels[split:]).astype(np.float64)
features_test = np.array(features[split:]).astype(np.float64)

print labels_test

classifier = LogisticRegression()
classifier.fit(features_train,labels_train)
print classifier.score(features_train,labels_train)

predictions = classifier.predict(features_test)


print predictions
print accuracy_score (labels_test, predictions)
```

```python
test_frame = pandas.read_csv('test.csv')
test_frame = test_frame.drop(['Cabin','Embarked','Ticket', 'Name', 'Age'], axis=1)
test_frame.dropna()
```

```python
print test_frame.to_string()
```

```python
features_main_test = []
for index, row in test_frame.iterrows():
    X = []
    
    X.append(row['PassengerId'])
    X.append(row['Pclass'])
    
    if row['Sex'] == 'male':
        X.append(0)
    else:
        X.append(1)
    X.append(row['SibSp'])
    X.append(row['Parch'])
    
    if not math.isnan(row['Fare']):
        X.append(row['Fare'])
    else:
        X.append(0.0)
    
    features_main_test.append(X)
len (features_main_test)
```

```python
features_main = np.array(features_main_test).astype(np.float64)

survived_predictions = classifier.predict(features_main)
len(survived_predictions)
```

```python
with open('predictions.csv', 'wb') as csvfile:
    fieldnames = ['PassengerId', 'Survived']
    writer = csv.DictWriter(csvfile, fieldnames=fieldnames)
    writer.writeheader()
    
    #write the data.
    for index in range(0, len(survived_predictions)):
        writer.writerow({'PassengerId': features_main_test[index][0] , 'Survived': survived_predictions[index]})
    
```
